```python
# Warning: this was a sprint job and there are no comments so far. 

import pandas as pd
from ecommercetools import seo

alphabet = "abcdefghijklmnopqrstuvxyz"
preposition = array = ["for", "to"]
df2 = pd.DataFrame(columns=["term", "relevance"])

for prep in preposition: 
    for letter in alphabet:
        expression = "how to use chatgpt " + prep + " " + letter
        results = seo.google_autocomplete(expression, include_expanded=False)
        df = pd.DataFrame(results, columns=["term", "relevance"])
        df2 = df2.append(df, ignore_index=True)

df2 = df2.sort_values(by='relevance', ascending=False)

df2.head(25)
```

```python
# Removing the generic chatbot


df2 = df2[df2['term'].str.contains('chatgpt') | df2['term'].str.contains('chat gpt')]

df2.head(250)
```
